```python
df.Product_Category_3 
```

```python
# First lets just try dropping column

df_temp = df.drop("Product_Category_3",axis=1,inplace=False)

df_temp.info()
```

```python
df_temp.isnull().sum()
```

```python
# Lets drop it row wise

df_temp[df_temp.Product_Category_2.isnull()].index
```

```python
df_temp = df_temp.drop(df_temp[df_temp.Product_Category_2.isnull()].index,axis=0,inplace=False)
```

```python
df_temp.info()
```

```python
df_temp.isnull().sum()
```

```python
# Now lets impute the missing values with bfill and ffill

df = df.fillna(method="ffill")

df.info()
```

```python
df.isnull().sum()
```

```python
# Above we can see that one one value is still missing because we have used forward filling , so the last value has nothing in forward.

#to rectify that we can do back filling do backward filling

df = df.fillna(method="backfill")

df.info()
```

```python
df.isnull().sum()
```
